## 🔍 R Nedir?

R, istatistiksel hesaplama ve veri analizi için tasarlanmış, fonksiyonel paradigmaya yakın çalışan, vektör tabanlı bir programlama dilidir. Özellikle veri biliminde "first-class citizen" olarak kabul edilir.

---

## ⚙️ R'nin Programlama Modeli

* **Vektör tabanlıdır** → Tüm işlemler genelde vektörler üzerinde yapılır
* **Fonksiyon-first yaklaşımı** → Her şey fonksiyonlarla yürür
* **Dinamik tip sistemi** → Tip belirtmeden değişken tanımlanır
* **İfade odaklıdır** → Her şey bir expression olarak değerlendirilir

# 🧪 R: Vektör ve Colon (:) Operatörü Mini Örnek

---

## 📦 Vektör  Tanımlama

R'de birden fazla sayıyı tek bir nesne içinde tutmak için **vektör** kullanılır.

# Vektör tanımlama

## ➕ Vektör ile Matematik İşlemleri

R vektör tabanlı çalıştığı için işlemler eleman bazlı yapılır.

```r
sayilar + 2
sayilar - 1
```

📌 Sonuç:

* 3 + 2 = 5
* 4 + 2 = 6
* 5 + 2 = 7

---

## 📌 Colon (:) Operatörü ile Vektör Oluşturma

`:` operatörü ardışık sayılar üretir.

```r
a <- 1:3   # 1 2 3

```

---

## ➕ Colon ile Matematik İşlemleri

```r

# toplama

# çıkarma

```

📌 Açıklama:

* İşlemler index bazlı yapılır
* Her eleman kendi karşılığı ile işlem görür

---

## 🔁 Örnek Çıktı Mantığı

```r
1:3 + 1:3
# [1] 2 4 6

1:5 - 1:5
# [1] 0 0 0 0 0
```

---

## 🔁 Farklı Uzunlukta Vektörlerle İşlem (Recycling Rule)

R, farklı uzunluktaki vektörlerde **recycling rule** (tekrar etme kuralı) uygular. Kısa vektör, uzun vektöre tamamlanacak şekilde baştan tekrar edilir.

```r
a <- 1:4   # 1 2 3 4
b <- 1:6   # 1 2 3 4 5 6
```

---

## 📊 Adım Adım Eşleşme Tablosu

| Index | a (1:4) | b (1:6) | a + b | a - b |
| ----: | ------: | ------: | ----: | ----: |
|     1 |       1 |       1 |     2 |     0 |
|     2 |       2 |       2 |     4 |     0 |
|     3 |       3 |       3 |     6 |     0 |
|     4 |       4 |       4 |     8 |     0 |
|     5 |       1 |       5 |     6 |    -4 |
|     6 |       2 |       6 |     8 |    -4 |

---

## ⚙️ R Davranış Mantığı

* `a` vektörü: 1 2 3 4 → tekrar eder → 1 2 3 4 1 2
* `b` vektörü: 1 2 3 4 5 6

### ➕ Toplama Sonucu

```r
a + b
# [1] 2 4 6 8 6 8
```

### ➖ Çıkarma Sonucu

```r
a - b
# [1] 0 0 0 0 -4 -4
```

⚠️ Not:

* Eğer uzunluk tam bölünmüyorsa R warning üretir
* Bu davranış "silent recycling bug" riskine yol açabilir

---

## 🎯 Özet

* `c()` → manuel vektör oluşturur
* `:` → ardışık sayı üretir
* R tüm işlemleri vectorized yapar (loop gerekmez)
* Toplama ve çıkarma eleman bazlı çalışır


# 🎲 R’de `sample()` Fonksiyonu

`sample()` fonksiyonu, R’de **rastgele örnekleme (random sampling)** işlemleri için kullanılır. Veri analizi, istatistik ve makine öğrenmesinde sıkça kullanılır.

---

## 🔧 Temel Kullanım

```r
sample(x, size, replace = FALSE)

Parametreler:
- x → Örnek alınacak veri (vektör veya sayı aralığı)
- size → Kaç eleman seçileceği
- replace → Tekrar seçime izin (TRUE / FALSE)

````python
x <- c(10, 20, 30, 40, 50)
sample(x, 3)
👉 1 ile 10 arasından rastgele 5 sayı seçer.
````

sample(1:5, 10, replace = TRUE)
👉 Aynı eleman birden fazla kez seçilebilir.
