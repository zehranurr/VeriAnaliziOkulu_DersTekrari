# 📊 R Programlama Dili Rehberi 

---

## 🔍 R Nedir?

R, istatistiksel hesaplama ve veri analizi için tasarlanmış, fonksiyonel paradigmaya yakın çalışan, vektör tabanlı bir programlama dilidir. Özellikle veri biliminde "first-class citizen" olarak kabul edilir.

---

## 🎯 Kullanım Alanları

* Veri analizi & veri bilimi
* İstatistiksel modelleme
* Makine öğrenmesi
* Finansal modelleme
* Akademik araştırmalar
* Veri görselleştirme

---

## ⚙️ R'nin Programlama Modeli

* **Vektör tabanlıdır** → Tüm işlemler genelde vektörler üzerinde yapılır
* **Fonksiyon-first yaklaşımı** → Her şey fonksiyonlarla yürür
* **Dinamik tip sistemi** → Tip belirtmeden değişken tanımlanır
* **İfade odaklıdır** → Her şey bir expression olarak değerlendirilir

## 💬 Yorum Satırları (Comments)

R'de yorum satırları `#` ile başlar:

```r
# Bu bir yorum satırıdır
x <- 10  # inline comment
```

## 🔢 Temel Veri Tipleri

```r
x <- 10          # numeric (double)
y <- 10L         # integer
z <- "merhaba"  # character
b <- TRUE        # logical
```

## 📦 Temel Veri Yapıları

```r
c(1,2,3)        # vector
list(1, "a")   # list
data.frame(x=1:3, y=c("a","b","c"))
matrix(1:6, nrow=2)
```

## ➗ Matematiksel Operatörler

```r
a <- 10
b <- 3

a + b   # toplama
a - b   # çıkarma
a * b   # çarpma
a / b   # bölme
a ^ b   # üs alma
a %% b  # mod (kalan)
a %/% b # tam bölme
```

---

## 📌 Colon (:) Operatörü

R'de `:` operatörü ardışık sayılar üretir:

```r
1:5      # 1 2 3 4 5
5:1      # 5 4 3 2 1
-2:2     # -2 -1 0 1 2
```

📌 Çok önemli: R vektör üretiminde en sık kullanılan yapıdır.

## 📐 Built-in Matematik Fonksiyonları

```r
sqrt(16)     # karekök
log(10)      # logaritma
exp(2)       # e^x
mean(c(1,2,3))
sum(c(1,2,3))
length(c(1,2,3))
```

## 🧠 Fonksiyon Tanımlama

```r
topla <- function(a, b) {
  a + b
}

selamla <- function(isim = "kullanıcı") {
  paste("Merhaba", isim)
}
```

## 🔁 Kontrol Yapıları

### If-Else

```r
x <- 10

if (x > 5) {
  print("büyük")
} else {
  print("küçük")
}
```

### For Loop (colon ile)

```r
for (i in 1:5) {
  print(i)
}
```

### While Loop

```r
i <- 1
while (i <= 5) {
  print(i)
  i <- i + 1
}
```

---

## 📦 apply Ailesi (Vectorization Alternatifi)

```r
x <- c(1,2,3,4)

lapply(x, function(i) i * 2)
sapply(x, function(i) i * 2)
```

## 📊 Data Frame İşlemleri

```r
df <- data.frame(
  isim = c("Ali", "Ayşe"),
  yas = c(25, 30)
)

# kolon erişimi
df$isim

# filtreleme
df[df$yas > 26, ]
```

## 🔍 Indexing Mantığı

```r
x <- c(10,20,30,40)

x[1]     # ilk eleman
x[2:4]   # aralık
x[-1]    # ilk eleman hariç
```
