# 📊 R Programlama Dili Rehberi

---

## 🔍 R Nedir?

R, istatistiksel hesaplama ve veri analizi için geliştirilmiş açık kaynaklı bir programlama dilidir. Özellikle veri bilimciler, istatistikçiler ve akademisyenler tarafından yaygın olarak kullanılır.

R dili, veri manipülasyonu, istatistiksel modelleme ve görselleştirme konularında güçlü kütüphaneler sunar.

---

## 🎯 R Nerelerde Kullanılır?

R dili aşağıdaki alanlarda yoğun olarak tercih edilir:

* 📈 Veri analizi ve veri bilimi
* 📊 İstatistiksel modelleme
* 🤖 Makine öğrenmesi
* 🧬 Biyoenformatik
* 💰 Finansal analiz
* 📉 Veri görselleştirme (grafikler, dashboardlar)

---

## ⚙️ R Dilinin Temel Özellikleri

* Açık kaynaklıdır (ücretsiz)
* Güçlü istatistiksel fonksiyonlara sahiptir
* Zengin paket (library) ekosistemi vardır
* Veri görselleştirme için çok güçlü araçlar sunar (ggplot2 gibi)
* Vektör bazlı çalışır (Python’dan farklı olarak)

---

## 🧠 R vs Python (Kısa Karşılaştırma)

| Özellik        | R                         | Python             |
| -------------- | ------------------------- | ------------------ |
| Amaç           | İstatistik & veri analizi | Genel amaçlı       |
| Öğrenme eğrisi | Orta                      | Kolay-Orta         |
| Kütüphaneler   | İstatistik ağırlıklı      | Çok geniş          |
| Kullanım alanı | Akademik & araştırma      | Endüstri & yazılım |

---

## 💻 RStudio Nedir?

RStudio, R dili için geliştirilmiş bir IDE'dir (Integrated Development Environment).

R kodlarını yazmayı, çalıştırmayı ve analiz etmeyi kolaylaştırır.

### 🔧 RStudio Özellikleri:

* Kod editörü
* Konsol (kod çalıştırma alanı)
* Grafik ve çıktı görüntüleme
* Paket yönetimi
* Proje yönetimi

---

## 🧪 Basit R Kod Örneği

```r
# Vektör oluşturma
sayilar <- c(1, 2, 3, 4, 5)

# Ortalama hesaplama
ortalama <- mean(sayilar)

print(ortalama)
```

---

## 📦 Popüler R Paketleri

* **ggplot2** → Veri görselleştirme
* **dplyr** → Veri manipülasyonu
* **tidyr** → Veri düzenleme
* **caret** → Makine öğrenmesi
* **shiny** → Web uygulamaları

---

## 🚀 R ile Neler Yapabilirsin?

* Veri setlerini analiz edebilirsin
* İstatistiksel testler uygulayabilirsin
* Grafikler ve dashboardlar oluşturabilirsin
* Makine öğrenmesi modelleri geliştirebilirsin
* Akademik çalışmalar için analizler yapabilirsin

---

## 📌 Sonuç

R, özellikle veri analizi ve istatistiksel işlemler için optimize edilmiş güçlü bir programlama dilidir. Eğer veri bilimi veya akademik analiz alanına yönelmek istiyorsan öğrenilmesi oldukça değerlidir.

---

## 🎯 Ek Not

Eğer Python biliyorsan, R öğrenmek veri analizi bakış açını ciddi şekilde güçlendirir. Özellikle istatistiksel modelleme tarafında R çok daha derin araçlar sunar.
