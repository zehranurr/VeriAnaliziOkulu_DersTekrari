Python’da import (içe aktarma) işlemi, hazır modülleri, paketleri veya kütüphaneleri kendi kod dosyanın içine dahil ederek kullanabilmeni sağlar. Bu sayede sıfırdan yazmak yerine, önceden geliştirilmiş güçlü araçlardan faydalanırsın.

Bu işlem import anahtar kelimesi ile yapılır ve genellikle kısaltma (alias) kullanılarak daha okunabilir hale getirilir.

```python
import numpy as np              # Sayısal hesaplamalar ve matris işlemleri
import pandas as pd             # Veri manipülasyonu ve analiz
import matplotlib.pyplot as plt # Veri görselleştirme ve grafik çizimi
```

Python’da import math ifadesi, yerleşik math (matematik) modülünü programa dahil eder. Bu modül; karekök, üs alma, logaritma gibi birçok temel matematiksel fonksiyonu hazır olarak sunar.
math nedir?

math modülü Python’un standart kütüphanesinde yer alır ve ek kurulum gerektirmeden kullanılabilir. Matematiksel işlemleri daha hızlı ve doğru şekilde yapmanı sağlar.

📌 math. içindeki nokta ne anlama gelir?

math. ifadesindeki nokta (.), modülün içindeki fonksiyonlara veya değişkenlere erişmek için kullanılır.

```python
import math

math.sqrt(16)   # 4.0 -> karekök alma
math.pi         # 3.14159... -> pi sayısı
math.factorial(5)  # 120 -> faktöriyel hesaplama
```

📦 Kütüphaneler ve Paket Yönetimi

Python’da projeler genellikle yalnızca standart modüllerle sınırlı değildir. Veri analizi, web scraping, doğal dil işleme gibi alanlarda kullanılan birçok harici kütüphane bulunur. Bu kütüphaneleri projeye dahil etmek için import yapmadan önce sistemimize kurmamız gerekir.

Bu kurulum işlemi Python’un paket yöneticisi olan pip ile gerçekleştirilir.

⚙️ pip Nedir?

pip (Python Package Installer), Python için geliştirilmiş paketleri indirip kurmamızı ve yönetmemizi sağlayan araçtır.
Kısaca, Python ekosistemindeki “uygulama mağazası” gibi çalışır.

💻 Paket Kurulumu

Bir paketi kurmak için terminal (komut satırı) kullanılır:

Windows: CMD (Komut İstemi)
Mac / Linux: Terminal

Kurulum komutu şu şekildedir:

- pip install pandas
- pip install matplotlib
- pip install beautifulsoup4
- pip install nltk

📌 Yaygın Kullanılan Kütüphaneler
- pandas: Veri analizi ve tablo işlemleri
- matplotlib: Grafik ve veri görselleştirme
- beautifulsoup4: Web sayfalarından veri çekme (web scraping)
- nltk: Doğal dil işleme (NLP) araçları
