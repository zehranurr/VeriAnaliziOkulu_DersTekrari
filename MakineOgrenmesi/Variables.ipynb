## 📦 Değişkenler (Variables)

Değişkenler, programlama sırasında veriyi **geçici olarak saklamak** için kullanılan isimlendirilmiş bellek alanlarıdır.  
Bir değeri hafızada tutmanıza, gerektiğinde bu değeri güncellemenize ve programın farklı noktalarında tekrar kullanmanıza olanak sağlar.

Başka bir deyişle, değişkenler sayesinde:
- Verileri **tekrar tekrar yazmak zorunda kalmazsınız**
- Kodunuz daha **okunabilir ve sürdürülebilir** olur
- Program akışı içinde **dinamik işlemler** gerçekleştirebilirsiniz

---

## 🧾 Standart Veri Tipleri (Built-in Types)

Python’da her değişken, aslında belirli bir **sınıfın (class)** örneğidir.  
Yani bir değişkene değer atadığınız anda, Python o verinin **tipini otomatik olarak belirler**.

Bu yaklaşım sayesinde Python, **dinamik tipli (dynamic typing)** bir dil olarak sınıflandırılır.

---

## 🧠 Temel Veri Tipleri

En sık kullanılan yerleşik (built-in) veri tipleri:

| Veri Tipi | Açıklama            | Örnek          |
|----------|--------------------|---------------|
| `bool`   | Mantıksal değer     | `True`, `False` |
| `int`    | Tamsayı             | `5`           |
| `float`  | Ondalıklı sayı      | `5.27836532`  |
| `str`    | Metin (string)      | `"Hello World"` |

---

## 💡 Örnek Kullanım

```python
my_int = 10            # int (tamsayı)

my_string = "Merhaba"  # str (karakter dizisi)

````python
my_string = "Merhaba"
print(type(my_string))
````

## 🔧 Fonksiyonların Yapısı

Python'da fonksiyonlar, belirli bir işi tekrar kullanılabilir şekilde tanımlamak için kullanılır.

### 📌 Temel Bileşenler

* **Fonksiyon Tanımı (Header)**
  `def` anahtar kelimesi ile başlar. Ardından fonksiyon adı ve parametreler yazılır.

* **Açıklama Metni (Docstring)**
  Fonksiyonun ne yaptığını anlatan açıklamadır. Üç tırnak (`"""`) ile yazılır.

* **Fonksiyon İçeriği (Body)**
  İşlemlerin gerçekleştiği kısımdır.

* **Geri Dönüş (return)**
  Fonksiyon sonucunu döndürür. Eğer kullanılmazsa `None` döner.

---

### ✅ Örnek: İki Sayıyı Toplama

```python
def topla(x, y):
    """İki değerin toplamını döndürür"""
    return x + y

print(topla(5, 9))
```

## 🧠 Alıştırmalar

### 1️⃣ Üç Sayının En Büyüğünü Bulma

```python
def en_buyuk(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

print(en_buyuk(4, 12, 7))
```

---

### 2️⃣ Listedeki Sayıların Çarpımı

```python
def carpim_hesapla(liste):
    sonuc = 1
    for sayi in liste:
        sonuc *= sayi
    return sonuc

veriler = [2, 3, 4, -2, 5]
print(carpim_hesapla(veriler))
```

### 3️⃣ Benzersiz Elemanları Bulma

#### Yöntem 1 (Set kullanımı)

```python
def unique_list(liste):
    return list(set(liste))
```

#### Yöntem 2 (Manuel kontrol)

```python
def unique_manual(liste):
    sonuc = []
    for eleman in liste:
        if eleman not in sonuc:
            sonuc.append(eleman)
    return sonuc

ornek = [1,1,2,2,3,4,4,5]
print(unique_manual(ornek))
```

---

### 4️⃣ Asal Sayı Kontrolü

```python
def asal_kontrol(n):
    if not isinstance(n, int) or n <= 1:
        return "Geçersiz"

    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(asal_kontrol(11))
```

## 📁 Dosya İşlemleri

### ✍️ Dosyaya Yazma

```python
dosya_adi = "metin.txt"

with open(dosya_adi, "w") as f:
    f.write("Kısa hikaye:\n")
    f.write("Satılık bebek ayakkabısı, hiç giyilmedi.")
```

---

### 📖 Dosya Okuma

```python

    icerik = f.read()

print(icerik)
```

### 🔍 Satırlara Ayırma

---

### 📊 Kelime Sayısı Hesaplama

## 🎯 Notlar

* `with open(...)` kullanımı, dosyayı otomatik kapatır.
* `set()` veri yapısı tekrar eden elemanları kaldırır.
* Asal sayı kontrolünde performans için `sqrt(n)` optimizasyonu kullanılır.
