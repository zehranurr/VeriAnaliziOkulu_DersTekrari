## 🖨️ Ekrana Yazdırma (print)

Python’da ekrana çıktı vermek için kullanılan temel fonksiyon **`print()`** fonksiyonudur.

Bu fonksiyon, parantez içine yazılan değeri (metin, sayı veya değişken) ekrana yazdırır. Python’da metinler hem tek tırnak (`' '`) hem de çift tırnak (`" "`) ile yazılabilir.

---





````python
### 🧪 Örnek: İlk Python çıktımız
print("Hello World")
````

## 💬 Yorum Satırları (Comments)

Python’da kod yazarken, yazdığımız satırların ne yaptığını açıklamak için **yorum satırları** kullanırız.

Yorumlar, program tarafından **çalıştırılmaz**; sadece kodu okuyan kişiye açıklama yapmak amacıyla kullanılır. Bu sayede kod hem daha anlaşılır hem de daha sürdürülebilir hale gelir.

---

### 🔹 Yorum Satırı Nasıl Yazılır?

Python’da yorum satırları `#` sembolü ile başlar:

```python
# Bu bir yorum satırıdır
print("Merhaba Dünya")  # Bu kısım açıklama içindir
